```python
import numpy as np
import matplotlib.pyplot as plt

# 
def sigmoid_like(x, N, k=10):
    return 1 - 1 / (1 + np.exp(-k * (x - N / 2)))

# 
def polynomial_like(x, N, p=2):
    return (1 - x / N) ** p

def step_function(x, N, m):
    # 
    w = N / m
    # 
    return 1 - np.floor(x / w) * (1 / m)

# 
N = 1000
x = np.linspace(0, N, 500)

# 
y_sigmoid = sigmoid_like(x, N, k=5)
y_polynomial = polynomial_like(x, N, p=2)
y_step = step_function(x, N, m=50)

# 
plt.figure(figsize=(8, 6))
plt.plot(x, y_sigmoid, label="Sigmoid-like Function", color="blue")
plt.plot(x, y_polynomial, label="Polynomial-like Function", color="orange")
plt.plot(x, y_step, label="Step Function", color="green")
plt.title("Custom Function Design")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
